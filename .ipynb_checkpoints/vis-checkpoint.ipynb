```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
first_set = pd.read_csv("dataset1.csv")
```

```python
first_set.head()
```

```python
menus = ["toolglass", "flowmenu", "toolpalette", "controlmenu"]

figures, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

menu_index = 0
for i in range(2):
    for j in range(2):
        sns.distplot(first_set[first_set["menu"] == menus[menu_index]]["time"], ax=axes[i][j])
        axes[i][j].set_title(menus[menu_index])
        menu_index += 1

```
